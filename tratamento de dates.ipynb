### Criações de Datas

```python
# Python - date
data_python = date(2023, 7, 25) # ano, mês, dia

display(data_python)
print(data_python)
```

```python
# Python
data_hora_python = datetime(2023, 7, 25,  15, 30) #Ano,mês,dia,hora,minuto

display(data_hora_python)
print(data_hora_python)
```

```python
type(data_hora_python)
```

```python
data_hora_python.strftime('%d/%m/%y') # jeito brasileiro
```

```python
data_hora_python.strftime('%Y-%m-%y') # internacional
```

```python
data_hora_python.strftime('%m/%d/%Y') # U.S - Estilo vale do silício
```

```python
data_hora_python.strftime('Today is %m/%d/%Y')
```

```python
# Lembrando que 'f' é de format e 'p' é de parse
data_hora_python.strftime("Today is %B %dth %Y")
```

```python
# Python - Time
hora_python = time(15,30) # Hora, minuto

display(hora_python)
print(hora_python)
```

```python
type(hora_python)
```

```python
hora_python.strftime("%H horas e %M minutos") # str FORMAT time strftime
```

```python
# String de data e hora
data_hora_str = '2023-07-25 15:30:00'

# Conversão da string para datetime ('p' de PARSING)
data_hora_2 = datetime.strptime(data_hora_str, '%Y-%m-%d %H:%M:%S')

display(data_hora_2)
print(data_hora_2)
```

### Conversão em STRING

```python
# Criação de um time
hora = time(15, 30)

# Formatação para string
hora_string = hora.strftime('%H:%M:%S')

print(f'Hora como string: {hora_string}')
```

### Cálculo de diferenças entre DATAS

```python
data1 = datetime(2023, 7, 25)
data2 = datetime(2023, 7, 30)

print(f'A primeira data é: {data1}\nA segunda data é: {data2}')
```

```python
data1, data2
```

```python
diferenca = data2 - data1 # TimeDelta

diferenca
```

```python
print(f'A diferença entre datas: {diferenca.days} em dias') # Acessando apenas dias no TimeDelta
```

```python
print(f'A diferença entre datas: {diferenca.seconds} em segundos')
```

```python
# Numero de segundos em 5 dias
diferenca.total_seconds()
```

### Formatação de DATAS

```python
# Python - DateTime
data_formatada_python = data_hora_python.isoformat()
print(f'Data formatada Python datetime (ISO 8601): {data_formatada_python}')
```

### 'E qual a diferença entre elas?'

```python
# Criando doi objetos datetime
data3 = datetime(2023, 7, 25, 14, 22, 28)
data_formatada = data3.isoformat()
data4 = datetime(2025, 3, 13, 1, 33, 33)

print(f'{data3}\n{data4}\n{data_formatada}')
```

```python
# Calculando a diferença entre as datas:
difenca_2 = data4 - data3
diferenca_3 = data3 - data4

print(f'{difenca_2}\n{diferenca_3}')
display(difenca_2)
display(diferenca_3)
```

```python
# Diferença em segundos:
segundos = difenca_2.total_seconds()
print(f'Diferença em segundos: {segundos:,}')
```

```python
# Diferença em anos (considerando um ano como 365.25 dias para incluir anos bissextos)
dias = difenca_2.days
print(f'Diferença em dias é: {dias}')
```

```python
# Diferença em minutos:
dif_minutos = segundos / 60
print(f'Diferença em dias é: {dif_minutos:,.2f}')
```

```python
# Diferença em horas:
dif_horas = dif_minutos / 60
print(f'Diferença em dias é: {dif_horas:,.2f}')
```

```python
# Diferença em anos (considerando um ano como 365.25 dias para incluir anos bissextos):
dif_anos = dias / 365.25
print(f'Diferença em dias é: {dif_anos:,.2f}')
```
